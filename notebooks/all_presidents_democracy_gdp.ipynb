## Analysing the visited countries by democracy and gdp levels

```python
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#data on presidential traveling as in other notebooks
df=pd.read_csv("../data/All data_geopanda_names2.csv", header=0,encoding = 'unicode_escape', parse_dates=["Date start", "Date end"])
```

```python
df.head()
```

## Visited countries and democracy

The data used here comes from the World Democracy Index by the Economist. The data exists from 2006 (except the years 2007 and 2009)

```python
df['Year'] = df['Date start'].apply(lambda x: x.year)
```

```python
#taking into ccount only the years after 2005
df1=df[df['Year']>2005].copy()
```

```python
df1.head()
```

```python
#reading in the data on democracy index 2006-2019
democracy_all=pd.read_csv("../data/Democracy Index all years_1.csv")
```

```python
democracy_all.head()
```

```python
democracy_years=pd.merge(df1, democracy_all, left_on=['Location country', "Year"], right_on=['Country', 'Year'])
```

```python
democracy_years
```

## China and democracy

```python
china_dem_all=democracy_years[democracy_years['Country of president']=='China'].copy()
```

```python
china_dem_all['democracy_index'].describe()
#the average democracy index is 5.96
```

```python
china_dem_all.sort_values(by='democracy_index').head(10)[["Country of president", "Location country", "Year", "democracy_index"]]
```

## Russia

```python
Russia_dem_all=democracy_years[democracy_years['Country of president']=='Russia'].copy()
```

```python
Russia_dem_all['democracy_index'].describe()
#the average is only 5.4. It seems it is because of many visits to Tajikistan, Turkmenistan and Uzbekistan
```

```python
Russia_dem_all.sort_values(by='democracy_index').head(10)[["Country of president", "Location country", "Year", "democracy_index"]]
```

## USA

```python
USA_dem_all=democracy_years[democracy_years['Country of president']=='USA'].copy()
```

```python
USA_dem_all['democracy_index'].describe()
#US has by far the highest mean of democracy index
```

```python
USA_dem_all.sort_values(by='democracy_index').head(10)[["Country of president", "Location country", "Year", "democracy_index"]]
#USa has  only two countries under 2
```

```python
democracy_years[democracy_years['democracy_index']<2].groupby("Country of president")["Location country"].nunique()
#US has 2 countries, China 6, Russia 6
```

## The different presidents of the US?

```python
presidents.groupby(["President"])["democracy_index"].mean()
#There is no significant difference between the means. 
#On the other hand, Obama traveled more to Africa where the index tends to be lower.
```

## GDP

```python
#using UN data 2017 on gdps as in unvisited countries
df_ranking=pd.read_csv("../data/gdp2017_ranking_UN.csv")
```

```python
world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))
```

## GDP rankings

```python
# one way to compare the countries is to see, how they rank among themselves based on gdps. 
```

```python
gdp_ranking2017_per_capita=pd.merge(df_ranking, world_counts, left_on='name', right_on='Location country')
```

```python
gdp_ranking2017_per_capita.head()
```

```python
df_gdp_norm_ranking=pd.read_csv("../data/gdp3.csv")
```

```python
df_gdp_norm_ranking
```

```python
gdp_UN_rankings=pd.merge(gdp_ranking2017_per_capita, df_gdp_norm_ranking, left_on='Location country', right_on='Country/Territory')
```

```python
#checking that small countries are in
gdp_UN_rankings[gdp_UN_rankings['Country/Territory']== 'Malta']
```

## Add in GDP data

```python
country_counts = df.groupby(["Country of president", "Location country"])["Order"]\
    .size()\
    .to_frame()\
    .reset_index()\
    .rename(columns={"Order": "num_trips"})
```

```python
country_counts.head()
```

```python
gdp_country_rankings = pd.merge(country_counts, df_gdp_norm_ranking, left_on='Location country', right_on='Country/Territory')
```

```python
gdp_country_rankings_all=pd.merge(gdp_country_rankings, df_ranking, left_on='Country/Territory', right_on='name')
```

```python
gdp_country_rankings_all.head()
```

```python
#median of rank at nominal gdp in the world of target countries by China
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "China") & (x["num_trips"]>1)
]["Rank/wholegdp"].median()
```

```python
#median of rank at nominal gdp in the world of target countries by Russia
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "Russia") & (x["num_trips"]>1)
]["Rank/wholegdp"].median()
```

```python
#median of rank at nominal gdp in the world of target countries by the US
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "USA") & (x["num_trips"]>1)
]["Rank/wholegdp"].median()
```

```python
#median of rank at gdp per capita in the world of target countries by  China
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "China") & (x["num_trips"]>1)
]["Rank"].median()
```

```python
#median of rank at gdp per capita in the world of target countries by Russia
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "Russia") & (x["num_trips"]>1)
]["Rank"].median()
```

```python
#median of rank at gdp per capita in the world of target countries by the US
gdp_country_rankings_all.loc[
    lambda x: (x["Country of president"] == "USA") & (x["num_trips"]>1)
]["Rank"].median()
```

USA has visited richer countries in both per capita and per nominal gdp than China and Russia when looking at the median rankings. China and Russia are closer to each other.
