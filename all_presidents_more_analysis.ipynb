# China and gdp

```python
China=df[df["Country of president"]=='China']
```

```python
China
```

```python
China.dtypes
```

```python
china_counts = China["Location country"].value_counts().to_frame().reset_index()
```

```python
#creating dataframe with china and visited countries, their gdps and their rankings by nominal gdp
China_rankings=pd.merge(china_counts, df_gdp_norm_ranking, left_on='index', right_on='Country/Territory')
```

```python
#adding gdp per capita and ranking per gdp/capita
China_rankings_all=pd.merge(China_rankings, df_ranking, left_on='Country/Territory', right_on='name')
```

```python
#checking how the most visited countries of China rank by the nominal gdp (median ranks of the top 40 countries visited)
China_rankings_all['Rank/wholegdp'].head(40).median()
```
